###Numerical Attributes

*Central Tendency*

min,max,midrange,mode,median

```python
col = 'Views'
min = df[col].min()
max = df[col].max()
median = df[col].median()
mode = df[col].mode()[0]
midrange = (max - min)/2
print('col:',col,
      '\n\tmin:', min,
      'max:',max,
      'median:', median,
      'mode:', mode,
      'midrange:', midrange)
```

```python
def central_tendency(col):
  min = df[col].min()
  max = df[col].max()
  median = df[col].median()
  mode = df[col].mode()[0]
  midrange = (max - min)/2
  print('col:',col,
        '\n\tmin:', min,
        '\n\tmax:',max,
        '\n\tmedian:', median,
        '\n\tmode:', mode,
        '\n\tmidrange:', midrange)
numerical_cols=['Views','Likes','Comments','Stream']
for col in numerical_cols:
  central_tendency(col)
```

*Dispersion*

range,quantiles,var,std



```python
col = 'Views'
range = df[col].max() - df[col].min()
quantiles = df[col].quantile([0.25, 0.5, 0.75])
IQR = quantiles[0.75] - quantiles[0.25]
var = df[col].var()
std = df[col].std()
print('col:',col,
      '\n\trange:', range,
      '\n\tQ1:',quantiles[0.25],
      '\n\tQ2:', quantiles[0.5],
      '\n\tQ3:', quantiles[0.75],
      '\n\tIQR:', IQR,
      '\n\tvar:', var,
      '\n\tstd:', std)
```

```python
def get_dispersion(col):
  range = df[col].max() - df[col].min()
  quantiles = df[col].quantile([0.25, 0.5, 0.75])
  IQR = quantiles[0.75] - quantiles[0.25]
  var = df[col].var()
  std = df[col].std()
  print('col:',col,
      '\n\trange:', range,
      '\n\tQ1:',quantiles[0.25],
      '\n\tQ2:', quantiles[0.5],
      '\n\tQ3:', quantiles[0.75],
      '\n\tIQR:', IQR,
      '\n\tvar:', var,
      '\n\tstd:', std)
numerical_cols=['Views','Likes','Comments','Stream']
for col in numerical_cols:
  get_dispersion(col)
```
